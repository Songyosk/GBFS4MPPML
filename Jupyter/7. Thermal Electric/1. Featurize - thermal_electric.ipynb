#### Merge results

```python
# Import packages
import pandas as pd
import joblib

from pymatgen import core
```

```python
# Import custom packages
import sys
sys.path.insert(1, 'C:/Users/Songy/OneDrive/Documents/GBSFS4MPP_local/new_properties')

import featurizers as feat
```

```python
# Load Materials object file
df1 = joblib.load('boltztrap_mp_composition.pkl')
df2 = joblib.load('boltztrap_mp_structural.pkl')
df3 = joblib.load('boltztrap_mp_structural_jarviscfid.pkl')
df4 = joblib.load('boltztrap_mp_additional_features.pkl')
```

```python
# Concat dataframes
frames = [df1, df2, df3, df4]

df_merged = pd.concat(frames, axis=1)

# # Col to drop
# cols = ['formula', 'structure', 'composition', 'oxidation_structure', 'oxidation_composition']
cols = ['formula', 'structure', 'composition', 'oxidation_composition']

df_merged = df_merged.drop(cols, axis=1)

# Nan to zero
df_merged = df_merged.fillna(0)

# False and True -> 0 and 1
df_merged.replace({False: 0, True: 1}, inplace=True)
df_merged
```

```python
# Initialization of featurizer
descriptors = feat.prepare_to_featurize(df=df_merged)

# One hot encoding of categorical features
df_merged, categorical_feature = descriptors.OHE()

categorical_feature
```

```python
# Load Materials object file
df_original = joblib.load('boltztrap_mp.pkl')
df_original.head()
```

```python
# Load Materials object file
df_original = joblib.load('boltztrap_mp.pkl')
df_original = df_original[['pf_n', 'pf_p', 's_n', 's_p', 'm_n', 'm_p']]

# Concat dataframes
frames = [df_merged, df_original]

df_merged = pd.concat(frames, axis=1)
df_merged
```

```python
# Save dataset 
joblib.dump(df_merged, 'boltztrap_mp_merged.pkl')
```
