### Module for treating an imbalanced dataset for classification ('imbalanced_dataset.py')

```python
import imbalanced_dataset as im
```

```python
# Path to the data file and path to save the new data file
path_to_file = '/Users/songyojung/data/df_train.pkl'
path_to_save = '/Users/songyojung/retrieved_data/AB3C4D6'
```

```python
# Initialize with data file and location to save new data file
# Ensure last column of the data file is the target variable or class
dataset = im.imbalanced_dataset(path_to_file, path_to_save)
```

```python
# Oversample e.g. using random oversampling with Gaussian noise (i.e. 'smoothed_ros') 
dataset.oversampling(
                    stratify = True,
                    technique = 'smoothed_ros'
                    )
```

```python
# Scale the features (optional step)
dataset.scale_features()
```

```python
# Save the Oversampled & original training sets, along with the test set
dataset.save(name = 'data', csv = False)
```
