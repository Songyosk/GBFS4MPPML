### Load data

```python
import pandas as pd  
import joblib

from monty.serialization import loadfn
```

```python
data1 = loadfn("backup/magnetic-orderings-database-snapshot.part1.json.gz")
data2 = loadfn("backup/magnetic-orderings-database-snapshot.part2.json.gz")
```

```python
df1 = pd.json_normalize(data1)
df2 = pd.json_normalize(data2)

df = pd.concat([df1, df2])
df
```

```python
len(df.columns)
```

```python
df.to_csv('raw_mag_order_data.csv')
joblib.dump(df, 'raw_mag_order_data.pkl')
```

```python
# Filter on stable structure 
df_stable = df.loc[df['stable'] == True]
df_stable
```

```python
joblib.dump(df_stable, 'mag_order_data_stable.pkl')
```

```python
joblib.dump(df, 'mag_order_data.pkl')
```
