#### Merge results

```python
# Import packages
import pandas as pd
import joblib
import re

from pymatgen import core
```

```python
# Import custom packages
import sys
sys.path.insert(1, 'C:/Users/Songy/OneDrive/Documents/GBSFS4MPP_local/new_properties')

import featurizers as feat
```

```python
# Load Materials object file
cols = ['formula', 'structure', 'composition', 'oxidation_structure', 'oxidation_composition']

df1 = joblib.load('mag_order_composition.pkl')
df1 = df1.reset_index(drop=True)
df1 = df1.drop(cols, axis=1)

df2 = joblib.load('mag_order_structural.pkl')
df2 = df2.reset_index(drop=True)
df2 = df2.drop(cols, axis=1)

df3 = joblib.load('mag_order_structural_jarviscfid.pkl')
df3 = df3.reset_index(drop=True)

df4 = joblib.load('mag_order_additional_features.pkl')
df4 = df4.reset_index(drop=True)
df4 = df4.drop(cols, axis=1)

df_target = joblib.load('mag_order_data_stable.pkl')
df_target = df_target[['energy_per_atom', 'energy_above_ground_state_per_atom', 'energy_diff_relax_static', 'total_magnetization', 'ordering']]
df_target = df_target.reset_index(drop=True)
```

```python
# Concat dataframes
frames = [df1, df2, df3, df4, df_target]

df_merged = pd.concat(frames, axis=1)
mag_order = pd.DataFrame({'mag_order': df_target['ordering'].tolist()})
df_merged = df_merged.join(mag_order)
df_merged
```

```python
# Remove duplications
df_merged  = df_merged.loc[:,~df_merged.columns.duplicated()]

# Remove space and JSON character in the column
df_merged.columns = df_merged.columns.str.replace(' ','_')
df_merged = df_merged.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', '', x))

# False and True -> 0 and 1
df_merged.replace({False: 0, True: 1}, inplace=True)
df_merged
```

```python
# Initialization of featurizer
descriptors = feat.prepare_to_featurize(df=df_merged)

# One hot encoding of categorical features
df_merged, categorical_feature = descriptors.OHE(col='mag_order')

categorical_feature
```

```python
df_merged = descriptors.movecol(['energy_per_atom', 'energy_above_ground_state_per_atom', 'energy_diff_relax_static', 'total_magnetization', 'mag_order'], ref_col='ordering_ohe_NM')
df_merged.head()
```
